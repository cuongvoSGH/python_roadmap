# Functions and Modules - Practice Notebook

Learn to write reusable functions and use built-in Python modules for common tasks.

```python
# Example: Define a simple function
def calculate_tax(amount, tax_rate):
    """Calculate tax on an amount"""
    tax = amount * tax_rate
    return tax

# Use the function
invoice_amount = 1000
tax_amount = calculate_tax(invoice_amount, 0.1)
print(f"Amount: ${invoice_amount}")
print(f"Tax: ${tax_amount}")

# Example: Using datetime module
import datetime
today = datetime.date.today()
print(f"Today's date: {today}")

# Example: Using math module
import math
result = math.ceil(99.1)
print(f"Ceiling of 99.1: {result}")
```
