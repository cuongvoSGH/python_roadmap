## 1. If Statements

Use if to execute code when a condition is true.

```python
# Example: Check invoice amount
invoice_amount = 5000

if invoice_amount > 1000:
    print("Large invoice - requires approval")

if invoice_amount > 0:
    print("Valid invoice amount")
    
# Example: Check account status
account_balance = 500
minimum_balance = 1000

if account_balance < minimum_balance:
    print("Warning: Low balance")
```

## 2. If-Else Statements

Handle both true and false conditions.

```python
# Example: Check payment status
is_paid = True

if is_paid:
    print("Invoice has been paid")
else:
    print("Invoice is still pending")

# Example: Validate amount
amount = 0

if amount > 0:
    print("Valid amount")
else:
    print("Invalid: amount must be positive")
```

## 3. If-Elif-Else Statements

Check multiple conditions in order.

```python
# Example: Categorize invoice amount
amount = 500

if amount >= 10000:
    print("Very large invoice - C-level approval required")
elif amount >= 5000:
    print("Large invoice - Manager approval required")
elif amount >= 1000:
    print("Medium invoice - Team lead approval required")
else:
    print("Small invoice - Approved automatically")

# Example: Account status check
balance = 0

if balance > 1000:
    print("Healthy balance")
elif balance > 0:
    print("Low balance - consider depositing")
else:
    print("Overdrawn - immediate action needed")
```
