## 1. Integer, Float, and String Types

Different data types for different kinds of information.

```python
# Example: Different data types
invoice_number = 1001  # Integer
total_amount = 2500.50  # Float (ALWAYS use float for money)
customer_name = "ABC Corporation"  # String
account_active = True  # Boolean

# Display with type information
print(f"Invoice: {invoice_number} - Type: {type(invoice_number)}")
print(f"Amount: {total_amount} - Type: {type(total_amount)}")
print(f"Customer: {customer_name} - Type: {type(customer_name)}")
print(f"Active: {account_active} - Type: {type(account_active)}")
```

## 2. Type Conversion

Converting between data types is essential for handling financial data from different sources.

```python
# Example: Type conversion
# Reading data from user input (always comes as string)
amount_text = "2500.50"
amount_float = float(amount_text)
print(f"Text: {amount_text}, Type: {type(amount_text)}")
print(f"Float: {amount_float}, Type: {type(amount_float)}")

# Converting float to string
price = 99.99
price_string = str(price)
print(f"String: {price_string}, Type: {type(price_string)}")

# Converting string to integer
quantity_text = "10"
quantity_int = int(quantity_text)
print(f"Integer: {quantity_int}, Type: {type(quantity_int)}")
```

## 3. Best Practices for Financial Data

**IMPORTANT:** Always use float for financial amounts, never int.

```python
# WRONG: Using int for money
wrong_amount = 100  # This loses decimal places

# RIGHT: Using float for money
correct_amount = 100.00  # Preserves decimal places

# Example of why this matters
price = 99.99
# If stored as int: 99 (wrong!)
# If stored as float: 99.99 (correct!)

# Calculate total with tax
item_price = 100.00  # Always float
tax_rate = 0.10
total_with_tax = item_price * (1 + tax_rate)
print(f"Item: ${item_price}")
print(f"Tax rate: {tax_rate * 100}%")
print(f"Total: ${total_with_tax}")
```
