```python
# Example: List comprehension - filter positive amounts
amounts = [100, -50, 200, -30, 300]
positive_amounts = [amount for amount in amounts if amount > 0]
print(f"Positive amounts: {positive_amounts}")

# Example: List comprehension - transform (calculate tax on each)
invoice_amounts = [100, 200, 300]
tax_on_each = [amount * 0.1 for amount in invoice_amounts]
print(f"Tax per invoice: {tax_on_each}")

# Example: Error handling
try:
    user_input = "not a number"
    amount = float(user_input)
except ValueError:
    print("Error: Invalid amount, must be a number")

# Example: File writing
with open("test_report.txt", "w") as file:
    file.write("Account Summary\n")
    file.write("Balance: $5000.00\n")
```
