# Operations - Practice Notebook

Learn to access and manipulate data structures using indexing, slicing, and membership tests.

```python
# Indexing: access elements by position
amounts = [100, 200, 300, 400, 500]
print(f"First: {amounts[0]}")
print(f"Last: {amounts[-1]}")

# Slicing: get multiple elements
recent_amounts = amounts[2:5]
print(f"Recent: {recent_amounts}")

# Dictionary access
invoice = {"id": 1001, "amount": 2500, "status": "paid"}
print(f"Invoice ID: {invoice['id']}")
print(f"Amount: ${invoice['amount']}")

# Membership test
departments = ("Finance", "Accounting", "HR")
if "Finance" in departments:
    print("Finance department found")
```
