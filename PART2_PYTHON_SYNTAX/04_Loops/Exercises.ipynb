# Loops - Practice Notebook

Learn to repeat code using for and while loops. Process multiple invoices and transactions efficiently.

```python
# Example: For loop with a list of amounts
invoice_amounts = [1000, 2500, 500, 3000]

print("Invoice Processing:")
for amount in invoice_amounts:
    print(f"Processing invoice: ${amount}")

# Example: For loop with range
print("\nCounting:")
for i in range(3):
    print(f"Number: {i}")

# Example: While loop with counter
print("\nWhile loop example:")
count = 0
while count < 3:
    print(f"Count: {count}")
    count += 1
```
