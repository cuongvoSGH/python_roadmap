# Data Structures - Practice Notebook

Learn to organize data using lists, tuples, dictionaries, and sets. Choose the right structure for each task.

```python
# Example: List - ordered, changeable
invoice_ids = [1001, 1002, 1003, 1004]
print("List:", invoice_ids)

# Example: Tuple - ordered, unchangeable
months = ("January", "February", "March", "April")
print("Tuple:", months)

# Example: Dictionary - key-value pairs (records)
customer = {
    "id": 5001,
    "name": "ABC Corp",
    "balance": 5000.00,
    "active": True
}
print("Dictionary:", customer)

# Example: Set - unique values only
department_codes = {"FIN", "ACC", "SAL", "FIN"}
print("Set:", department_codes)  # Note: duplicate "FIN" removed
```
