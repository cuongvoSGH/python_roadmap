# Reconciliation Workflows — Practices: merges, indicators, mismatches

This notebook walks through common reconciliation steps used in finance: outer merges, indicator checks, tolerance-based mismatch detection, and summarising exceptions.

```python
import pandas as pd
import numpy as np

# sample ledgers and bank payments
ledger = pd.DataFrame({'invoice':[1001,1002,1003,1004],'amount':[120.0,200.0,150.0,300.0]})
bank = pd.DataFrame({'invoice':[1001,1003,1005],'paid':[120.0,149.5,50.0]})

# standard reconciliation outer merge
m = pd.merge(ledger, bank, on='invoice', how='outer', indicator=True, suffixes=('_led','_bank'))
display(m)

# detect mismatches within tolerance (e.g., rounding differences)
matched = m[m['_merge']=='both']
mismatch = matched[~(matched['amount_led'].fillna(0).round(2) == matched['paid'].fillna(0).round(2))]
display(mismatch)

# summary counts by merge indicator
display(m['_merge'].value_counts())
```
