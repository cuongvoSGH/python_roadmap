# Excel Integration — Practices: read_excel, to_excel, openpyxl

Examples show reading and writing Excel files and simple tips for using `openpyxl` via pandas.

```python
import pandas as pd
import numpy as np

# create sample DataFrame and write to Excel (requires openpyxl installed)
df = pd.DataFrame({'invoice':[101,102,103], 'amount':[100,200,150], 'date':pd.to_datetime(['2024-01-01','2024-01-05','2024-01-10'])})
fn = 'sample_transactions.xlsx'
df.to_excel(fn, index=False)
print('Wrote', fn)
```

## Read Excel with parse_dates and usecols

```python
df2 = pd.read_excel(fn, parse_dates=['date'], usecols=['invoice','amount','date'])
display(df2)
```
