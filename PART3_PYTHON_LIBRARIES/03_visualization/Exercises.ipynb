# Visualization — Practices: matplotlib & seaborn

This notebook shows small examples for plotting common finance charts: time-series, bar, histogram, and scatter with seaborn styling.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
sns.set_theme(style='whitegrid')

# sample time-series data
dates = pd.date_range('2024-01-01', periods=12, freq='M')
df = pd.DataFrame({'date': dates, 'revenue': np.linspace(1000,2000,12) + np.random.randn(12)*100})
df.set_index('date', inplace=True)

# time-series plot
ax = df['revenue'].plot(title='Monthly revenue')
ax.set_ylabel('USD')
plt.show()
```

## Bar chart and histogram

```python
# bar chart: top customers
cust = pd.Series([500,800,200,1200], index=['A','B','C','D'])
cust.sort_values().plot(kind='barh', title='Top customers')
plt.xlabel('Amount')
plt.show()

# histogram: distribution of transaction amounts
amounts = np.random.exponential(200, size=500)
plt.hist(amounts, bins=30)
plt.title('Amount distribution')
plt.show()
```

## Scatter with regression (seaborn)

```python
x = np.random.randn(100) * 100 + 500
y = x * 0.2 + np.random.randn(100)*20
sns.regplot(x=x, y=y)
plt.title('Scatter with linear fit')
plt.xlabel('Metric X')
plt.ylabel('Metric Y')
plt.show()
```
