# NumPy — Practices: Arrays, Broadcasting, np.where, np.select

This notebook demonstrates core NumPy techniques used in finance workflows, and compares `np.where` with pandas alternatives.

```python
import numpy as np
import pandas as pd

# basic array creation
a = np.array([100, -5, 200, 0])
b = np.array([1,2,1,3])
a, b
```

## Broadcasting and vectorized ops

```python
# broadcasting example
prices = np.array([10.0, 20.0, 30.0])
quantities = np.array([[1],[2],[3]])
total = prices * quantities  # broadcasts prices across rows
total
```

## np.where examples and pandas interop

```python
arr = np.array([100, -5, 200, 0])
clean = np.where(arr > 0, arr, 0)
clean
# with pandas Series (index preserved via explicit Series constructor)
s = pd.Series(arr, index=['a','b','c','d'])
s_clean = pd.Series(np.where(s.gt(0), s, 0), index=s.index)
s_clean
```

## Multiple conditions: np.select vs nested np.where

```python
condlist = [arr < 0, (arr >=0) & (arr < 150), arr >= 150]
choicelist = [-1, 0, 1]
labels = np.select(condlist, choicelist, default=np.nan)
labels
```

## Performance note: avoid Python loops; use vectorized ops

```python
# build large arrays and time np.where vs list comprehension
import time
large = np.random.randn(1_000_000) * 1000
start = time.time()
res_np = np.where(large > 0, large, 0)
t_np = time.time() - start
start = time.time()
res_py = [x if x>0 else 0 for x in large]
t_py = time.time() - start
t_np, t_py
```
